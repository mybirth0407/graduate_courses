```python
from nltk.corpus import gutenberg   # Docs from project gutenberg.org
files_en = gutenberg.fileids()      # Get file ids
doc_en = gutenberg.open('austen-emma.txt').read()

sentences_en = nltk.tokenize.sent_tokenize(doc_en)

noun_adj_list = []
for sentence in sentences_en:
    tokens = nltk.tokenize.word_tokenize(sentence)
    tags = nltk.pos_tag(tokens)
    for word, tag in tags:
        if tag in ['NN', 'NNP', 'ADJ', 'JJ', 'VB', 'VBD']:
            noun_adj_list.append(word.lower())
```

```python
from nltk.stem import PorterStemmer

from nltk.corpus import stopwords
stop_words = stopwords.words('english') + [
    '.', ',', '--', '\'s', '?', '!', ')', '(', ':', ';', '\'', '[',']',
    '\'re', '"', '-', '}', '{', u'—', 'rt', 'http', 't', 'co', '@', '#',
    'mr.', 'mrs.']

stemmer=PorterStemmer()

stemmed = []
for token in noun_adj_list:
    if token == '' or token in stop_words:
        continue
    stemmed.append(stemmer.stem(token))

from collections import Counter

counts = Counter(stemmed)
tags = counts.most_common(300)
print(tags)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


word_dict = dict(tags)
print(word_dict)
```

```python
wordcloud = WordCloud(max_font_size=40, max_words=1000)
wordcloud.generate_from_frequencies(word_dict)

plt.figure(figsize = (10, 10), facecolor = None) 

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from collections import Counter
from konlpy.tag import Okt
okt = Okt()

pos_tokens = okt.pos(doc_ko)

noun_adj_list = []
for word, tag in pos_tokens:
    if tag in ['Noun', 'Adjective']:
        noun_adj_list.append(word)

#형태소별 count
counts = Counter(noun_adj_list)
tags = counts.most_common(300)
print(tags)
```

```python
font_path = '/usr/share/fonts/truetype/nanum/NanumSquareR.ttf'     # A font of your choice

wordcloud = WordCloud(font_path=font_path, max_font_size=100, max_words=1000, margin=2)

ko_dict = dict(tags)
wordcloud.generate_from_frequencies(ko_dict)

plt.figure(figsize = (8, 8), facecolor = None) 

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
