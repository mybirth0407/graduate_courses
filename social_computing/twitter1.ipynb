## 키워드 분석 1: 더 글로리

```python
df_glory = pd.read_json("twitter/the_glory.json")
```

```python
df_glory.head()
```

### 1. 언제 트윗되었나?

#### 가장 최근 트윗

```python
df_glory.head(1)
```

#### 가장 마지막 트윗

```python
df_glory.tail(1)
```

트윗의 시간을 통해 대략 48분 정도에 1100개의 트윗이 생성되었음을 알 수 있다.

#### 1100개 트윗의 평균 시각

```python
df_glory['tw_unixtime'] = df_glory['created_at'].apply(lambda x: x.timestamp())
dt = datetime.datetime.fromtimestamp(df_glory['tw_unixtime'].mean())
dt.ctime()
```

주어진 데이터의 트윗은 평균적으로 14시 54분즈음 이루어졌음을 알 수 있다.

### 2. 언제 계정을 만들었나?

```python
df_glory_user = pd.DataFrame(data=df_glory['user'].tolist())
```

```python
df_glory_user.head()
```

```python
df_glory_user[["screen_name", "created_at"]].head()
```

```python
df_glory_user['user_created_datetime'] = \
    df_glory_user['created_at'].apply(
        lambda x: datetime.datetime.strptime(x, '%a %b %d %H:%M:%S +%f %Y'))
df_glory_user.head()
```

#### 주어진 데이터 중 가장 빨리/최근에 만들어진 계정

```python
df_glory_user['user_created_unixtime'] = \
    df_glory_user['user_created_datetime'].apply(lambda x: x.timestamp())
```

```python
min_idx = df_glory_user[['user_created_unixtime']].idxmin()
df_glory_user.iloc[min_idx][["screen_name", "created_at"]]
```

```python
max_idx = df_glory_user[['user_created_unixtime']].idxmax()
df_glory_user.iloc[max_idx][["screen_name", "created_at"]]
```

#### 계정을 만든 시각의 평균 시각

```python
df_glory_user['user_created_unixtime'] = \
    df_glory_user['user_created_datetime'].apply(lambda x: x.timestamp())
dt = datetime.datetime.fromtimestamp(df_glory_user['user_created_unixtime'].mean())
dt.ctime()
```

계정을 만든 시각의 평균을 계산하면, 2019년 5월 4일 오후 10시 정도이다.

### 3. 몇개나 트윗했는가?

```python
df_glory_user[['screen_name', 'statuses_count']].head()
```

#### 트윗에 대한 통계량

```python
df_glory_user[['screen_name', 'statuses_count']].describe()
```

최소와 최대의 차이, 분산이 굉장히 큰 것으로 보아, 소수의 하드 유저가 존재함을 알 수 있다.

### 4. 수집된 tweet 중 가장 follower가 많은 사용자는?

```python
max_idx = df_glory_user['followers_count'].idxmax()
df_glory_user.iloc[max_idx][["screen_name", "followers_count"]]
```

tpstm301 유저가 20573개의 follower를 갖고 있다.

```python
df_glory_user[["screen_name", "followers_count"]].describe()
```

팔로워 수의 통계량을 분석하면, 최대 값이 굉장히 높은 것에 비해 평균값과 사분위 값은 작다.<br>
즉, 일부 소수의 유저만이 수많은 팔로워를 갖고 있음을 알 수 있다.

### 5. 가장 많이 나오는 hashtag는?

```python
df_glory_hashtags = pd.DataFrame(data=df_glory['entities'].tolist())
df_glory_hashtags.head()
```

```python
tags = []
for row in df_glory_hashtags['hashtags']:
    if len(row) != 0:
        for text in row:
            tags.append(text['text'])
```

```python
counter = Counter(tags)
```

```python
counter.most_common()[:10]
```

### 6. 수집된 트윗을 분석하여 가장 많이 retweet된 (중복이 안되는) 트윗 10개를 찾고 이들의 리트윗 횟수, 원작성자, 트윗내용 등을 정리

```python
# nan 예외처리를 위한 전처리 과정
df_glory_rt_list = df_glory['retweeted_status'].tolist()
list2df = []
for item in df_glory_rt_list:
    if type(item) == dict:
        list2df.append(item)
```

```python
df_glory_rt = pd.DataFrame(data=list2df)
```

```python
c = []
for name, group in df_glory_rt.groupby('id_str'):
    c.append([
        group.iloc[0]['retweet_count'],
        group.iloc[0]['user']['screen_name'],
        group.iloc[0]['text']])
```

```python
pd.DataFrame(
    columns=['retweet_count', 'origin_user', 'text'],
    data=sorted(c, key=lambda x: x[0], reverse=True)[:10])
```
